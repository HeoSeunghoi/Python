```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 256)
s = np.sin(x)
c = np.cos(x)
plt.plot(x,c)
plt.plot(x,s)
plt.xticks([-np.pi,-np.pi/2, 0, np.pi/2, np.pi],['$-\pi$','$-\pi/2$','0','$pi/2$','$\pi$'])

plt.yticks([-1, 0, +1])
plt.show()
```
