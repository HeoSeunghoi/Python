```python
%%writefile Student.py
#student class
class student(object):
    #student class의 data 저장
    def __init__(self, num, name, kor, eng, math):
        self._num = int(num)
        self._name = name
        self._kor = int(kor)
        self._eng = int(eng)
        self._math = int(math)
        self._sum = self._kor + self._eng + self._math
        self._avg = self._sum/3
        self._rank = 0
        
    #property를 사용해 객체의 데이터를 직접 접근할 수 있도록 선언(num, name, kor, eng, math, sum, avg, rank)
    @property
    def num(self):
        return self._num
    @property
    def name(self):
        return self._name
    @property
    def kor(self):
        return self._kor
    @property
    def eng(self):
        return self._eng
    @property
    def math(self):
        return self._math
    @property
    def total(self):
        return self._sum
    @property
    def avg(self):
        return self._avg
    @property
    def rank(self):
        return self._rank
    
    #rank를 저장해 주기 위한 method
    def setrank(self, rank):
        self._rank = rank
```

```python
%%writefile StudentGradingSystem.py
#Student파일로부터 student class import
from Student import student

class StudentGradingSystem:
    #StudentGradingSystem의 student list를 초기화
    def __init__(self):
        self._lst = []
        
    #새로운 student의 정보가 들어오면 list에 추가하기 위한 method
    def register_student(self, name):
        self._lst.append(name)
        
    #지금까지 저장된 student의 정보를 바탕으로 필요한 정보를 구해주는 method
    def process(self):
        #계산을 위해 필요한 데이터 초기화
        rank = 1
        same_score = 0
        self.total = 0
        self.totalavg = 0
        self.korTotal = 0
        self.engTotal = 0
        self.mathTotal = 0
        self.kor = 0
        self.eng = 0
        self.math = 0
        save_total = 0
        
        #list에 저장된 데이터를 바탕으로 student의 점수들의 합을 계산
        for i in range(len(self._lst)):
            self.total += self._lst[i].total
            self.korTotal += self._lst[i].kor
            self.engTotal += self._lst[i].eng
            self.mathTotal += self._lst[i].math
            
        #list에 저장된 student들의 과목별, 총점의 평균 계산
        self.kor = self.korTotal/len(self._lst)
        self.eng = self.engTotal/len(self._lst)
        self.math = self.mathTotal/len(self._lst)
        self.totalavg = self.total/len(self._lst)
        
        #list에 저장된 student들을 total을 기준으로 내림차순으로 정렬
        self._sortedList = sorted(self._lst, key=lambda student: student.total, reverse = True)
        
        #정렬된 sortedList에 rank를 계산 (동점은 같은 rank로 저장)
        for i in range(len(self._sortedList)):
            if rank == 1:
                self._sortedList[i].setrank(rank)
                save_total = self._sortedList[i].total
                rank += 1
            elif self._sortedList[i].total == save_total:
                self._sortedList[i].setrank(rank-1)
                same_score += 1
            else:
                rank += same_score
                self._sortedList[i].setrank(rank)
                save_total = self._sortedList[i].total
                rank += 1
                same_score = 0
                
    #student들의 정보를 출력
    @property
    def print_students(self):
        for i in self._sortedList:
            print("번호: %d, 이름: %s 국어: %d, 영어: %d, 수학: %d, 총점: %d, 평균: %.2f, 등수: %d"%(i.num, i.name, i.kor, i.eng, i.math, i.total, i.avg, i.rank))
        print()
            
    #class에 속한 학생들의 평균들을 출력
    @property
    def print_class_information(self):
        print("총점 반평균: %.2f\n국어 반평균: %.2f\n영어 반평균: %.2f\n수학 반평균: %.2f\n"%(self.totalavg, self.kor, self.eng, self.math))
```

```python
%%writefile midterm.py
#만들어준 2개의 파일로부터 student와 StudentGradingSystem을 import
from Student import student
from StudentGradingSystem import StudentGradingSystem

#실행시간을 구하기 위해 time을 import
import time

class verbose:
    def __init__(self, func):
        self.func = func
    def __call__(self):
        start_time = time.time()
        self.func()
        print("실행시간은: %.19lf"%(time.time()-start_time))

@verbose
def program_start():
    studentdata = StudentGradingSystem()
    Gang1 = student(1, "서강일", 90, 80, 85)
    studentdata.register_student(Gang1)
    Gang2 = student(2, "서강이", 90, 85, 95)
    studentdata.register_student(Gang2)
    Gang3 = student(3, "서강삼", 80, 80, 80)
    studentdata.register_student(Gang3)
    Gang4 = student(4, "서강사", 90, 92, 83)
    studentdata.register_student(Gang4)
    Gang5 = student(5, "서강오", 85, 90, 90)
    studentdata.register_student(Gang5)
    studentdata.process()
    studentdata.print_students
    studentdata.print_class_information

program_start()
```
