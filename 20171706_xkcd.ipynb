```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
with plt.xkcd():
    plt.title('XKCD style plot!!!')
    x = np.linspace(-np.pi, np.pi, 256)
    c = np.cos(x)
    plt.plot(x,c)
    plt.scatter([(np.pi * 2)/3],  [-0.5], color = "r", linewidth = 1)
    plt.annotate('0.5 Here!', xy = ((np.pi * 2)/3, -0.5), xytext = (-50, -50), xycoords = 'data', textcoords = 'offset points', fontsize=12, arrowprops=dict(arrowstyle = "->", linewidth=2))
    
plt.show()
```
